# **Dropping Duplicate Entries**

```python
df.drop_duplicates(inplace=True)
df
```

```python
df['country'] = df['country'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['director'] = df['director'].fillna('Unknown')

df
```

```python
df['date_added'] = pd.to_datetime(df['date_added'].astype(str).str.strip(), errors='coerce')
df
```

```python
df['year_added'] = df['date_added'].dt.year
df
```

#**Number of Movies vs TV Shows**

```python
sns.countplot(data=df, x='type', palette='Set2')
plt.title('Distribution of Content Types')
plt.show()
```

 # **Top 10 Countries by Content**

```python
top_countries = df['country'].value_counts().head(10)
sns.barplot(x=top_countries.values, y=top_countries.index, palette='muted')
plt.title('Top 10 Countries by Netflix Content')
plt.xlabel('Count')
plt.ylabel('Country')
plt.show()
```

# **Distrbution Based on the Content Released Per Year**

```python
df['year_added'].value_counts().sort_index().plot(kind='bar', color='lightblue')
plt.title('Content Added Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')
plt.show()
```

# **Top 10 Genres on Netflix**

```python
df['listed_in'].str.split(', ', expand=True).stack().value_counts().head(10).plot(kind='bar', color='teal')
plt.title('Top 10 Genres on Netflix')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()
```

# **World Cloud of Description**

```python
from wordcloud import WordCloud

text = ' '.join(df['description'].dropna().values)
wordcloud = WordCloud(width=800, height=800, background_color='black').generate(text)

plt.figure(figsize=(12,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Descriptions')
plt.show()
```
